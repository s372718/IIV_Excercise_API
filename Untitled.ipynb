## 1. REST APIs Grundlagen

### 1.1 Schnittstellen, Webdienste und REST APIs

Ein Webdienst ist eine Sammlung von offenen Protokollen und Standards, die für den Datenaustausch zwischen Anwendungen oder Systemen verwendet werden. Software-Anwendungen, die in verschiedenen Programmiersprachen ( z.B. Java und Python) geschrieben wurden und auf verschiedenen Plattformen (z.B. Windows und Linux) laufen, können Webdienste nutzen, um Daten über Computernetzwerke wie das Internet auszutauschen. Diese Interoperabilität ist auf die Verwendung offener Standards zurückzuführen.

Webdienste, die auf der REST-Architektur basieren, sind als RESTful-Webdienste bekannt. Die im folgenden betrachteten  RESTful-Webdienste gelten als minimalistisch aufgebaut, hoch skalierbar und einfach wartbar und werden deswegen sehr häufig zur Erstellung von APIs für webbasierte Anwendungen verwendet. REST steht für „Representational State Transfer“ und bedeutet, dass ein Dienst eindeutig adressierbare **Ressourcen, zustandslos** anbietet.

### 1.2 REST Daten und ihre Formate

In der REST-Architektur bietet ein REST-Server Zugriff auf Ressourcen, während ein REST-Client auf die Ressourcen zugreift oder diese modifiziert. Jede Ressource wird dabei durch URIs/ globale IDs identifiziert. REST verwendet verschiedene Formate zur Darstellung einer Ressource wie Text, JSON, XML. JSON ist dabei das wohl verbreitetste Format, da es neben den Rohen daten die Struktur diese fest mitgibt.

### 1.3 REST Schnittstellen bedienen

RESTful-Webdienste verwenden HTTP-Methoden, um das Konzept der REST-Architektur zu implementieren. Ein RESTful-Webdienst definiert in der Regel einen URI, einen Uniform Resource Identifier einen Dienst, bietet eine Ressourcendarstellung wie JSON und eine Reihe von HTTP-Methoden.

Die folgenden vier HTTP-Methoden werden üblicherweise in REST-basierten Architekturen verwendet.

- GET - Bietet einen Nur-Lese-Zugriff auf eine Ressource.

- POST - Wird verwendet, um eine neue Ressource zu erstellen.

- DELETE - Wird verwendet, um eine Ressource zu entfernen.

- PUT - Wird verwendet, um eine vorhandene Ressource zu aktualisieren oder eine neue Ressource zu erstellen.


Bei nicht REST-Schnittstellen, insbesondere bei Schnittstellen einer höherer Abstraktionsschicht, kann es sein, dass die REST Operationen von den sogenannten CRUD Operationen abgelöst werden. Dise stammen aus der Programmierung insbesondere zur Abfrage und Manipulation persistenter Speicher. 

- Create = Datensatz anlegen,
- Read oder Retrieve = Datensatz lesen,
- Update = Datensatz aktualisieren,
- Delete oder Destroy = Datensatz löschen.

### 1.6 Fehler erkennen

Statuscodes werden mit einer Antwort nach jedem Aufruf an den Server zurückgegeben. Sie beschreiben kurz das Ergebnis des Aufrufs. Es gibt eine große Anzahl von Statuscodes, folgende Beispiele zeigen einige der häufigeren Stuscodes.

- 200 - OK. Die Anfrage war erfolgreich. Die Antwort selbst hängt von der verwendeten Methode (GET, POST usw.) und der API-Spezifikation ab.
- 204 - Kein Inhalt. Der Server hat die Anfrage erfolgreich bearbeitet und keinen Inhalt zurückgegeben.
- 301 - Permanent verschoben. Der Server antwortet, dass die angeforderte Seite (Endpunkt) an eine andere Adresse verschoben wurde und leitet auf diese Adresse um.
- 400 - Der Server kann die Anfrage aufgrund von Fehlern auf der Client-Seite (falsches Anfrageformat) nicht verarbeiten.
- 401 - Nicht autorisiert. Tritt auf, wenn die Authentifizierung fehlgeschlagen ist, weil die Anmeldedaten falsch sind oder sogar fehlen.
- 403 - Verboten. Der Zugriff auf die angegebene Ressource wird verweigert.
- 404 - Nicht gefunden. Die angeforderte Ressource wurde auf dem Server nicht gefunden.
- 500 - Interner Server-Fehler. Tritt auf, wenn ein unbekannter Fehler auf dem Server aufgetreten ist.

Bemerkung: Aufgrund der Programmierung des APIs auf Serverseite im vorigen Beispiel wird, obwohl die Autorisierung nicht korrekt erfolgt ist, ein Statuscode 215 als Antwort vom API gegeben anstatt eines 401. Sowas sollte bei der Entwicklung einer API eigentlich vermieden werden. Da Statuscodes bei sehr vernetzten Systemen von der Aufrufenden Klasse ggf. je nach Statuscode weiter in einer Fehlerbehandlung behandelt werden können, um z.B. dem User nötige Informationen weiter zu geben oder ggf. eine alternative API aufzurufen.
